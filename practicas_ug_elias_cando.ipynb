```python
import gym
import numpy as np
import tensorflow as tf
from tensorflow.keras import models, layers

# Crear entorno CartPole
env = gym.make("CartPole-v1")
num_actions = env.action_space.n
num_states = env.observation_space.shape[0]

# Crear red neuronal para aproximar la función Q
model = models.Sequential([
    layers.Dense(24, activation='relu', input_shape=(num_states,)),
    layers.Dense(24, activation='relu'),
    layers.Dense(num_actions, activation='linear')
])

model.compile(optimizer='adam', loss='mse')

# Parámetros de entrenamiento
gamma = 0.95  # Factor de descuento
epsilon = 1.0  # Tasa de exploración inicial
epsilon_min = 0.01  # Tasa de exploración mínima
epsilon_decay = 0.995  # Factor de decaimiento de epsilon

# Entrenar el agente
for episode in range(100):
    state = env.reset()
    state = np.reshape(state, [1, num_states])
    done = False
    total_reward = 0

    while not done:
        # Explorar o explotar
        if np.random.rand() <= epsilon:
            action = np.random.choice(num_actions)  # Exploración
        else:
            action = np.argmax(model.predict(state)[0])  # Explotación

        next_state, reward, done, _ = env.step(action)
        next_state = np.reshape(next_state, [1, num_states])
        total_reward += reward

        # Actualización de Q (aprendizaje)
        target = reward
        if not done:
            target += gamma * np.max(model.predict(next_state)[0])
        target_f = model.predict(state)
        target_f[0][action] = target
        model.fit(state, target_f, epochs=1, verbose=0)

        state = next_state

    # Reducir epsilon (menos exploración con el tiempo)
    if epsilon > epsilon_min:
        epsilon *= epsilon_decay

    print(f'Episode {episode+1}: Total Reward = {total_reward}')
```

```python
import tensorflow as tf
from tensorflow.keras import layers

# Crear un modelo de red neuronal
model = tf.keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(4,)),
    layers.Dense(128, activation='relu'),
    layers.Dense(2, activation='softmax')  # 2 posibles acciones
])

# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Generar datos de ejemplo (aquí usas tus propios datos para entrenamiento)
import numpy as np
data = np.random.random((1000, 4))  # Ejemplo de datos de entrada
labels = np.random.randint(2, size=(1000,))  # Etiquetas

# Entrenar el agente
model.fit(data, labels, epochs=10)

# Predecir acciones
predictions = model.predict(np.array([[1.0, 0.5, -1.2, 0.3]]))
print(predictions)
```
